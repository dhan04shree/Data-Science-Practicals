```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the built-in Iris dataset from seaborn
iris = sns.load_dataset('iris')

# Display first few rows
print(iris.head())

# Display info about data types
print("\nFeature Types:")
print(iris.dtypes)
```

```python
# Histograms for each feature
iris.hist(figsize=(10, 8), bins=15, color='skyblue', edgecolor='black')
plt.suptitle('Feature Distributions - Iris Dataset')
plt.tight_layout()
plt.show()
```

```python
# Create boxplots for each numeric feature
numeric_features = iris.select_dtypes(include='float64').columns

plt.figure(figsize=(12, 6))
for i, col in enumerate(numeric_features, 1):
    plt.subplot(1, 4, i)
    sns.boxplot(y=iris[col], color='lightgreen')
    plt.title(col)
    plt.tight_layout()
plt.suptitle('Boxplots of Iris Features')
plt.tight_layout()
plt.show()
```

```python
# Boxplot grouped by species
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris, x='species', y='sepal_length')
plt.title('Sepal Length by Species')
plt.show()
```
